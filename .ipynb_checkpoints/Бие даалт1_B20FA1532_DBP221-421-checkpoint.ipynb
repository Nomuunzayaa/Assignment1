```python
import pandas as pd
import numpy as np
```

```python
#1 Ирц
a1 = pd.read_excel ('/Users/noza/Downloads/Term2-2/DBP221_421 A.Uuganbayar/Biy daalt/Бие даалт - 1.xlsx',sheet_name = ('Ирц'))
a1 = a1.dropna(axis = 1, how = 'all')
a1 = a1.fillna(0)
a1.replace(('и','т'),(1,0), inplace = True)
a2 = a1.drop(columns=['No'])
a2["Нийт"] = a2.sum(axis=1) * 10 // 32
a2
```

```python
#2 Семинар
a3 = pd.read_excel ('/Users/noza/Downloads/Term2-2/DBP221_421 A.Uuganbayar/Biy daalt/Бие даалт - 1.xlsx',sheet_name=('Семинар'))
a4 = a3.drop(columns=['No'])
a4["Нийт -25"] = a4.sum(axis=1) * 25 // 19
a4
```

```python
#3 Шалгалт
a5 = pd.read_excel ('/Users/noza/Downloads/Term2-2/DBP221_421 A.Uuganbayar/Biy daalt/Бие даалт - 1.xlsx',sheet_name=('Шалгалт'))
a6 = a5.drop(columns=['No'])
a6["Нийт"] = a6.sum(axis=1)
a6
```

```python
#4 Бие даалт
a7 = pd.read_excel ('/Users/noza/Downloads/Term2-2/DBP221_421 A.Uuganbayar/Biy daalt/Бие даалт - 1.xlsx',sheet_name=('Бие даалт'))
a8 = a7.drop(columns=['No'])
a8["Нийт"] = a8.sum(axis=1)
a8
```

```python
#5 Хүмүүжил хандлага
a9 = pd.read_excel ('/Users/noza/Downloads/Term2-2/DBP221_421 A.Uuganbayar/Biy daalt/Бие даалт - 1.xlsx',sheet_name=('Хүмүүжил хандлага'))
a10 = a9.drop(columns=['No'])
a10["Нийт"] = a10.sum(axis=1)
a10
```

```python
#6 Нэмэлт оноо
a11 = pd.read_excel ('/Users/noza/Downloads/Term2-2/DBP221_421 A.Uuganbayar/Biy daalt/Бие даалт - 1.xlsx',sheet_name=('Нэмэлт оноо'))
a12 = a11.drop(columns=['No'])
a12["Нийт"] = a12.sum(axis=1)
a12
```

```python
# Нийт
a13 = pd.read_excel ('/Users/noza/Downloads/Term2-2/DBP221_421 A.Uuganbayar/Biy daalt/Бие даалт - 1.xlsx',sheet_name=('Нийт'))
a13
```

```python
a14 = pd.DataFrame({'Оюутны нэр' : a13['Оюутны нэр'],
                    'Оюутны код' : a13['Оюутны код'],
                   'Ирц-10' : a2['Нийт'],
                   'Семинар-25' : a4['Нийт -25'],
                   'Шалгалт-30' : a6['Нийт'],
                   'Бие даалт-25' : a8['Нийт'],
                   'ХХ-10' : a10['Нийт'],
                   'Нэмэлт оноо' : a12['Нийт'],
                   'Хувь' : a13['Нийт']})
                   
a14["Хувь"] = a14.sum(axis = 1) 
a14
```

```python
# Үнэлгээ
a14['Үнэлгээ'] : a13['Unnamed: 11']
a14['Үнэлгээ']=np.where((a14['Хувь'] < 59.5 ),'F',
                  np.where((a14['Хувь'] >= 59.5 )&(a14['Хувь'] < 69.5 ),'D',
                  np.where((a14['Хувь'] >= 69.5 )&(a14['Хувь'] < 79.5 ),'C',
                  np.where((a14['Хувь'] >= 79.5 )&(a14['Хувь'] < 89.5 ),'B',
                  np.where((a14['Хувь'] >= 89.5 ),'A','NAN'
                          )))))
a14=a14.dropna(axis=1,how= 'all')
a14 = a14.dropna() 
a14
```

```python
# Тоолох
a14['Хэд'] = a13['Unnamed: 13']
a14['Аль'] = a13['Unnamed: 14']
a = a14.groupby(['Үнэлгээ'])['Үнэлгээ'].count()
for i in range(len(a)):
    a14['Хэд'][i]=a[i]
a14=a14.fillna(' ')
a14
```

```python
# Амжилт
b1 = a14['Үнэлгээ'].value_counts()['A']
b2 = a14['Үнэлгээ'].value_counts()['B']
b3 = a14['Үнэлгээ'].value_counts()['C']
b4 = a14['Үнэлгээ'].value_counts()['D']
b5 = a14['Үнэлгээ'].value_counts()['F']
b6 = (b1+b2+b3+b4) * 100 //(b1+b2+b3+b4+b5) 
print('Тухайн хичээлийн амжилт: ', b6, '%')
```

```python
# Чанар
b7 = (b1+b2) * 100 //(b1+b2+b3+b4+b5) 
print('Тухайн хичээлийн чанар: ', b7, '%')
```

```python
# Үнэлгээний арга тус бүр дэх хамгийн их утга
print('Хамгийн өндөр ирцийн оноо: ', a2.loc[:,['Нийт']].max())
print('Хамгийн өндөр семинарын оноо: ', a4.loc[:,['Нийт -25']].max())
print('Хамгийн өндөр шалгалтын оноо: ', a6.loc[:,['Нийт']].max())
print('Хамгийн өндөр бие даалт-1 оноо: ', a8.loc[:,['Бие даалт - 1']].max())
print('Хамгийн өндөр бие даалт-2 оноо: ', a8.loc[:,['Бие даалт - 2']].max())
print('Хамгийн өндөр ХХ-1 оноо: ', a10.loc[:,['ХХ-1']].max())
print('Хамгийн өндөр ХХ-2 оноо: ', a10.loc[:,['ХХ-2']].max())
print('Хамгийн өндөр ХХ-3 оноо: ', a10.loc[:,['ХХ-3']].max())
print('Хамгийн өндөр ХХ-4 оноо: ', a10.loc[:,['ХХ-4']].max())
print('Хамгийн өндөр нэмэлт оноо: ', a12.loc[:,['Нийт']].max())
```
